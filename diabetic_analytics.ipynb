```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the data present in the TSV file
df = pd.read_csv("diabetese.tsv" ,sep="\t") #data frame mae convert arne ke liye help hoga
df.head()

```

```python
df.info()
```

```python
#lets find the average age of diabetic patients
mean_age = df['AGE'].mean() #to find the central tendency (Avg kya ha)
print(f"Mean age is : {mean_age}")
```

```python
#variance
var_age = df['AGE'].var()
print(f" variance in Age : {var_age}")
```

```python
#standard deviation
std_age = df['AGE'].std()
print(f"Standard deviation is : {std_age}")
```

```python
#BoxPlotting for the age field
plt.figure(figsize=(2,6))
plt.boxplot(df['AGE'] , showmeans=True)
plt.grid(color='gray' , linestyle = "dotted")
plt.xlabel('Age')
plt.ylabel("years")
plt.show()
```

```python
#BoxPlotting for body mass index(BMI)

plt.figure(figsize = (6,2))
plt.boxplot(df['BMI'],vert=False, showmeans = True)
plt.grid(color = 'gray' , linestyle='dotted')
plt.xlabel("Value")
plt.ylabel("BMI")
plt.show()

```

```python
#BoxPlotting for Blood Pressure(BP)

plt.figure(figsize = (6,2))
plt.boxplot(df['BP'],vert=False, showmeans = True)
plt.grid(color = 'gray' , linestyle='dotted')
plt.xlabel("Value")
plt.ylabel("BP")
plt.show()
```

```python
#let understad what age group of the patients are the mostly diabetics

#we can create histogram for the same

df['AGE'].hist(bins=12)
plt.title("Age distribution of the diabetic patients")
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.show()
```

```python
#histogram for the body mass index(BMI)

df['BMI'].hist(bins=12)
plt.title("BMI distribution of the diabetic patients")
plt.xlabel('BMI Range')
plt.ylabel('Count')
plt.show()
```
